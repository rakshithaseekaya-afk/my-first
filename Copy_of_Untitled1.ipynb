```python
# Install dependencies as needed:
# !pip install kagglehub[pandas-datasets]
import os
import kagglehub
from kagglehub import KaggleDatasetAdapter

# 1. Download/load the dataset path
dataset_handle = "lalit7881/comprehensive-car-market-analysis"
path = kagglehub.dataset_download(dataset_handle)

# 2. Automatically find the CSV file in the dataset directory
files = os.listdir(path)
csv_files = [f for f in files if f.endswith('.csv')]

if csv_files:
    file_path = csv_files[0]
    print(f"Found dataset file: {file_path}")

    # 3. Load the dataframe using the updated dataset_load method
    df = kagglehub.dataset_load(
        KaggleDatasetAdapter.PANDAS,
        dataset_handle,
        file_path
    )

    print("\nFirst 5 records:")
    print(df.head())
else:
    print(f"No CSV files found in the dataset. Available files: {files}")
```
